```
# Import packages
import numpy as np
import pandas as pd
import altair as alt
```

```
# Load dataset
df = pd.read_csv('Anscombe.csv',sep=';')
df
```

```
I = alt.Chart(df[df.Category.eq('I')]).mark_circle(size=60).encode(
    alt.X('x',
        scale=alt.Scale(domain=(0, 20))
    ),
    alt.Y('y',
        scale=alt.Scale(domain=(0, 12))
    ),
    color='Category',
    tooltip=['Category', 'x', 'y']
)

boxplotx = I.mark_boxplot(size=100).encode(
    x='Category:O',
    y=alt.Y('x', scale=alt.Scale(domain=(0, 6)))
)

# Add regression line
final = alt.layer(
    I,
    I.transform_regression('x', 'y').mark_line()
)

final & boxplotx
```

```
# Build visualizations

# Create base charts
I = alt.Chart(df[df.Category.eq('I')]).mark_circle(size=60).encode(
    alt.X('x',
        scale=alt.Scale(domain=(0, 20))
    ),
    alt.Y('y',
        scale=alt.Scale(domain=(0, 12))
    ),
    color='Category',
    tooltip=['Category', 'x', 'y']
)

II = alt.Chart(df[df.Category.eq('II')]).mark_circle(size=60).encode(
    alt.X('x',
        scale=alt.Scale(domain=(0, 20))
    ),
    alt.Y('y',
        scale=alt.Scale(domain=(0, 12))
    ),
    color='Category',
    tooltip=['Category', 'x', 'y']
)

III = alt.Chart(df[df.Category.eq('III')]).mark_circle(size=60).encode(
    alt.X('x',
        scale=alt.Scale(domain=(0, 20))
    ),
    alt.Y('y',
        scale=alt.Scale(domain=(0, 12))
    ),
    color='Category',
    tooltip=['Category', 'x', 'y']
)

IV = alt.Chart(df[df.Category.eq('IV')]).mark_circle(size=60).encode(
    alt.X('x',
        scale=alt.Scale(domain=(0, 20))
    ),
    alt.Y('y',
        scale=alt.Scale(domain=(0, 12))
    ),
    color='Category',
    tooltip=['Category', 'x', 'y']
)

# Add regression line
chart1 = alt.layer(
    I,
    I.transform_regression('x', 'y').mark_line()
)

row1 = alt.hconcat(I, II)
row2 = alt.hconcat(III, IV)

alt.vconcat(row1, row2
).configure_axis(
    grid=False
).properties(
    title='Quarteto de Anscombe'
).configure_title(
    fontSize=23,
    anchor='middle'
)
```

```
# Selection tool
selection = alt.selection_single(fields=['Category'])
# Color change when clicked
color = alt.condition(selection,
                    alt.Color('Category:N'),
                    alt.value('lightgray'))

base = alt.Chart(df).mark_circle(size=60).encode(
    x='x',
    y='y',
    color='Category',
    tooltip=['Category', 'x', 'y']
)

# Chart
chart = base.encode(
  color=alt.condition(selection, 'Category:N', alt.value('lightgray')),
).add_selection(
  selection
)

final = base + chart 

final.properties(
    title='Quarteto de Anscombe'
).configure_title(
    fontSize=23,
    anchor='middle'
)
```

```
# Import packages
import numpy as np
import pandas as pd
import altair as alt
```

```
# Load dataset
df = pd.read_csv('Anscombe.csv',sep=';')
df
```

```
# Build visualizations

# Create base charts
I = alt.Chart(df[df.Category.eq('I')]).mark_circle(size=60).encode(
    alt.X('x',
        scale=alt.Scale(domain=(0, 20))
    ),
    alt.Y('y',
        scale=alt.Scale(domain=(0, 14))
    ),
    color='Category',
    tooltip=['Category', 'x', 'y']
)

II = alt.Chart(df[df.Category.eq('II')]).mark_circle(size=60).encode(
    alt.X('x',
        scale=alt.Scale(domain=(0, 20))
    ),
    alt.Y('y',
        scale=alt.Scale(domain=(0, 14))
    ),
    color='Category',
    tooltip=['Category', 'x', 'y']
)

III = alt.Chart(df[df.Category.eq('III')]).mark_circle(size=60).encode(
    alt.X('x',
        scale=alt.Scale(domain=(0, 20))
    ),
    alt.Y('y',
        scale=alt.Scale(domain=(0, 14))
    ),
    color='Category',
    tooltip=['Category', 'x', 'y']
)

IV = alt.Chart(df[df.Category.eq('IV')]).mark_circle(size=60).encode(
    alt.X('x',
        scale=alt.Scale(domain=(0, 20))
    ),
    alt.Y('y',
        scale=alt.Scale(domain=(0, 14))
    ),
    color='Category',
    tooltip=['Category', 'x', 'y']
)

# Add regression line
chartI = alt.layer(
    I,
    I.transform_regression('x', 'y', extent = (0,20)).mark_line()
)

chartII = alt.layer(
    II,
    II.transform_regression('x', 'y', extent = (0,20)).mark_line()
)

chartIII = alt.layer(
    III,
    III.transform_regression('x', 'y', extent = (0,20)).mark_line()
)

chartIV = alt.layer(
    IV,
    IV.transform_regression('x', 'y', extent = (0,20)).mark_line()
)

# Setting up all charts together

row1 = alt.hconcat(chartI, chartII)
row2 = alt.hconcat(chartIII, chartIV)

alt.vconcat(row1, row2
).configure_axis(
    grid=False
).properties(
    title='Quarteto de Anscombe'
).configure_title(
    fontSize=23,
    anchor='middle'
)
```

```
# Selection tool
selection = alt.selection_single(fields=['Category'])
# Color change when clicked
color = alt.condition(selection,
                    alt.Color('Category:N'),
                    alt.value('lightgray'))

base = alt.Chart(df).mark_circle(size=60).encode(
    x='x',
    y='y',
    color='Category',
    tooltip=['Category', 'x', 'y']
)

# Chart
chart = base.encode(
  color=alt.condition(selection, 'Category:N', alt.value('lightgray')),
).add_selection(
  selection
)

final = base + chart 

finalchart = final.properties(
    title='Quarteto de Anscombe'
).configure_title(
    fontSize=23,
    anchor='middle'
)

finalchart
```

```
# Configure the options common to all layers
brush = alt.selection_single(fields=['Category'])
base = alt.Chart(df).add_selection(brush)

# Configure the points
points = base.mark_circle(size=60).encode(
    x=alt.X('x', title=''),
    y=alt.Y('y', title=''),
    color=alt.condition(brush, 'Category', alt.value('lightgrey'))
)

# Configure the ticks
tick_axis = alt.Axis(labels=False, domain=False, ticks=False)

x_ticks = base.mark_tick().encode(
    alt.X('x', axis=tick_axis),
    alt.Y('Category', title='', axis=tick_axis),
    color=alt.condition(brush, 'Category', alt.value('lightgrey'))
)

y_ticks = base.mark_tick().encode(
    alt.X('Category', title='', axis=tick_axis),
    alt.Y('y', axis=tick_axis),
    color=alt.condition(brush, 'Category', alt.value('lightgrey'))
)

# Build the chart
finalcombo = y_ticks | (points & x_ticks)

finalcombo.properties(
    title='Quarteto de Anscombe'
).configure_title(
    fontSize=23,
    anchor='middle'
)
```
